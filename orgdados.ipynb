```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
results = pd.read_csv('results.csv')
goalscorers = pd.read_csv('goalscorers.csv')
shootouts = pd.read_csv('shootouts.csv')
```

```python
display(results)
display(shootouts)
```

```python
matches_in_country = results.groupby('country').size() # Representa o número de partidas ocorridas em cada país
matches_in_country_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(matches_in_country) # Destaca os países que sediam mais partidas

plt.figure(figsize=(10, 5))
plt.imshow(matches_in_country_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Países que mais sediam partidas", fontsize=16)
plt.show()
```

```python
matches_tournaments = results.groupby('tournament').size() # Número de partidas por torneio
matches_tournaments_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(matches_tournaments) # Destaca os torneios com mais partidas

plt.figure(figsize=(10, 5))
plt.imshow(matches_tournaments_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Torneios com mais partidas", fontsize=16)
plt.show()
```

```python
shootoutsWinners = shootouts.groupby('winner').size()
shootoutsWinners_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(shootoutsWinners)

plt.figure(figsize=(10, 5))
plt.imshow(matches_tournaments_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Maiores batedores de penâltis", fontsize=16)
plt.show()
```

```python
goals_by_team = goalscorers.groupby('team').size()
goals_by_team_wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(goals_by_team)

plt.figure(figsize=(10, 5))
plt.imshow(goals_by_team_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Países com mais gols", fontsize=16)
plt.show()
```

```python
best_scorers = goalscorers.groupby('scorer').size().sort_values(ascending=False).head(10).sort_values(ascending=True)

plt.figure(figsize=(10, 6))  # Ajusta o tamanho da figura
best_scorers.plot(kind='barh', color='skyblue')  # 'barh' cria barras horizontais

plt.title('Maiores Artilheiros')  # Título do gráfico
plt.xlabel('Contagem')  # Rótulo do eixo X (quantidade)
plt.ylabel('Jogadores')  # Rótulo do eixo Y (categorias)
plt.show()
```
